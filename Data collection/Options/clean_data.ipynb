```python
import pandas as pd
import json
import os
import glob

# Define input and output directories
input_folder = "."
output_folder = "Transformed_Spy_chains"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get a list of all CSV files in the input folder
csv_files = glob.glob(os.path.join(input_folder, "*.csv"))

# Parse the 'data' 
def parse_data_column(value):
    try:
        return json.loads(value.replace("'", "\""))
    except Exception as e:
        print(f"Error parsing value: {value}\nError: {e}")
        return {}

# Loop over each CSV file
for file in csv_files:
    print(f"Processing file: {file}")
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Apply the parse function to the 'data' column to create a new column of dictionaries
    df['data_dict'] = df['data'].apply(parse_data_column)
    
    # Expand the dictionaries into a DataFrame (each dictionary key becomes a column)
    data_expanded = pd.json_normalize(df['data_dict'])
    
    # Use the expanded DataFrame as the final DataFrame
    final_df = data_expanded.copy()
    
    # Create output file name by prefixing "transformed_" to the original file's base name
    base_name = os.path.basename(file)
    output_file = os.path.join(output_folder, f"transformed_{base_name}")
    
    # Save the final transformed DataFrame to a new CSV file
    final_df.to_csv(output_file, index=False)
    print(f"Saved transformed data to: {output_file}")
```

```python
# Get a list of all CSV files in the current directory
csv_files = glob.glob("./*.csv")

# List to collect each DataFrame
dfs = []

# Loop through each CSV file and read it
for file in csv_files:
    try:
        df = pd.read_csv(file)
        # Check if DataFrame is empty (no data rows)
        if not df.empty:
            dfs.append(df)
        else:
            print(f"Deleting empty file: {file}")
            os.remove(file)
        
    except pd.errors.EmptyDataError:
        # Handle case where the file has no content at all (not even headers)
        print(f"Deleting file with no content: {file}")
        os.remove(file)
    

# Merge all DataFrames into a single one
merged_df = pd.concat(dfs, ignore_index=True)

# Optional: Save the merged DataFrame to a new file
merged_df.to_csv("SPY_s Chain Barchart.csv", index=False)
print("âœ… Merged CSVs saved to 'merged_output.csv'")



```
